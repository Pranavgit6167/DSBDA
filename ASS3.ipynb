```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
df['Species'].unique()
```

```python
gp=df.groupby('Species')
```

```python
gp.groups.keys()
```

```python
gp.agg(['mean','median',pd.Series.mode,'min','max','var','std'])
```

```python
#there can be multiple mode for that us custom function
```

```python
def custom_mode(x):
    modes=x.mode()
    if len(modes)==len(x):
        return np.nan
    else:
        return modes.tolist()
aggregated_data = gp.agg(['mean', 'median',custom_mode, 'min', 'max', 'var', 'std'])
```
