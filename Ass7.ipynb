```python
Q1 = np.percentile(df, 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(df, 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
```

```python
upper=np.where(df>=(Q3+1.5*IQR))
lower=np.where(df<=(Q1-1.5*IQR))
```

```python
df.drop(upper[0], inplace=True)
df.drop(lower[0],inplace=True)
```
